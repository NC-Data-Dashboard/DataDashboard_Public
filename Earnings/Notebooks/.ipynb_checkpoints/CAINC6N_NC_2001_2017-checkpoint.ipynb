```python
import pandas as pd
import requests
from io import BytesIO, StringIO
from zipfile import ZipFile
```

```python
df_backup = pd.read_csv('./Updates/STG_BEA_CAINC6N_NC_2001_2017.txt')
df_backup.to_csv('./Backups/STG_BEA_CAINC6N_NC_2001_2017.txt)
```

```python
response = requests.get('https://apps.bea.gov/regional/zip/CAINC6N.zip')
zip_file = ZipFile(BytesIO(response.content))
files = zip_file.namelist()
with zip_file.open(files[34]) as csvfile:
    df = pd.read_csv(csvfile, encoding='ISO-8859-1', sep=",")
```

```python
df.tail(10)
```

```python
df_clean = df[:-3]
df_clean.tail(5)
```
